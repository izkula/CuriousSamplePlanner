```python
import os
import torch
import numpy as np
# First, get all of the files
from os import listdir
from os.path import isfile, join

import matplotlib.pyplot as plt
import sklearn
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
%matplotlib inline


# Constants
pathname = "../data_collection/solution_data/"
exp_name = "trajs_threeblocks"

path = pathname+exp_name
data_files = [f for f in listdir(path) if isfile(join(path, f))]

total_length = 0
traj_states = []
traj_actions = []
traj_rewards = []
traj_lens = []

g_states = []


print("Num Files: "+str(len(data_files)))
for file in data_files:
    data = torch.load(pathname+exp_name+"/"+file)
    g_states.append(torch.squeeze(torch.tensor(data['states']), dim=0))	

catted_states = torch.cat(g_states,0).numpy()
print(catted_states.shape)

# Project into 2 dimensions
pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=3))])

results = pipeline.fit_transform(catted_states)
print(results.shape)
plt.scatter(results[:1000,0], results[:1000,1], results[:1000,2])
plt.show()
```
